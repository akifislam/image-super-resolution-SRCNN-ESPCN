```python
import os
import cv2

# ---------- Paths ----------
input_folder = "/home/akif/ASG15/akif_photos"
output_folder = "/home/akif/ASG15/akif_photos_LR"

os.makedirs(output_folder, exist_ok=True)

# ---------- Choose Low-Resolution Size ----------
LR_width  = 1024
LR_height = 1024

# ---------- Process All Images ----------
for filename in os.listdir(input_folder):

    ext = os.path.splitext(filename)[1].lower()

    if ext in [".jpg", ".jpeg", ".png", ".jfif"]:

        input_path = os.path.join(input_folder, filename)

        # read image
        img = cv2.imread(input_path)
        if img is None:
            print("Could not read:", filename)
            continue

        # resize to LR
        lr_img = cv2.resize(img, (LR_width, LR_height), interpolation=cv2.INTER_CUBIC)

        # convert output filename to PNG
        base_name = os.path.splitext(filename)[0]
        output_path = os.path.join(output_folder, f"LR_{base_name}.png")

        # save
        cv2.imwrite(output_path, lr_img)

        print("Saved:", output_path)

print("ðŸŽ‰ All low-resolution PNG images created successfully!")
```
